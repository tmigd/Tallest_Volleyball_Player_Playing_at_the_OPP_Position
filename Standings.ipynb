```python
#Importing all the necessary packages
%load_ext autoreload
%autoreload 2

import pandas as pd # library for data analysis
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
#All the yearly standings for each japanese teams were found from the website:
#https://www.oddsportal.com/volleyball/japan/v-league/standings/

#All the yearly standings for each south korean teams were found from the website:
#https://www.oddsportal.com/volleyball/south-korea/volleyball-league/standings/
```

```python
#Wolfdogs Nagoya (Bartosz Kurek, 2020-)

season = ["2012-2013","2013-2014","2014-2015","2015-2016","2016-2017","2017-2018","2018-2019","2019-2020","2020-2021",
        "2021-2022","2022-Present",]
position = [5,5,3,1,2,2,5,7,3,2,1]
df=pd.DataFrame({"Season":season, "Position":position})

x_values = np.array(season)
y_values = np.array(position)
#plt.scatter(x_values, y_values)
plt.plot(x_values, y_values,marker="X", markersize=10, markerfacecolor="red",markeredgecolor="black")

plt.ylim(0.5,10)
plt.yticks(range(1,11,1))

plt.gca().invert_yaxis() #To invert the y-axis

plt.xticks(rotation=45)

plt.grid(linestyle = '--')

plt.title(label="Wolfdogs Nagoya's yearly standings, from last 10 years")
plt.xlabel('Seasons (as of 2022-11-24)') 
plt.ylabel('Position Rankings') 

plt.arrow(8, 4.5, 0, -.9,width=.08,color="black")
plt.text(x=6.5, y=5, s="Bartosz's 1st season",weight="bold")

plt.show()
```
